```python
# Opera dataframes e tabelas
import pandas as pd

# Trabalha com arrays e operações matemáticas
import numpy as np

# Biblioteca para construir e treinar redes neurais
import tensorflow as tf

# Biblioteca principal para criar modelos de redes neurais em Keras
from tensorflow import keras

# Define um modelo sequencial (camadas empilhadas)
from tensorflow.keras.models import Sequential

# Define camadas densas (totalmente conectadas) para redes neurais
from tensorflow.keras.layers import Dense, Dropout

# Separa os dados em conjuntos de treinamento e teste
from sklearn.model_selection import train_test_split

# Normaliza os dados para melhorar o desempenho da rede neural
from sklearn.preprocessing import StandardScaler

# Pacote para medir a performance do modelo com métricas como erro médio absoluto (MAE)
from sklearn.metrics import mean_absolute_error
```

## Criando um modelo de rede neural

### Carregando os dados

```python
# Carregar o dataset
df = pd.read_csv("/content/BIN-DataSet-Higiene.csv")
```

### Separando os dados

```python
# Selecionar as features e o target (score)
features = [
    "morning_brush", "afternoon_brush", "night_brush",
    "morning_floss", "afternoon_floss", "night_floss",
    "morning_mouthwash", "afternoon_mouthwash", "night_mouthwash",
    "total_brushes", "total_floss", "total_mouthwash", "streak_days"
]
X = df[features]
y = df["score"]
```

```python
# Normalizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividindo em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### Criando o modelo

```python
# Selecionar apenas as features necessárias
features_simplificadas = ["total_brushes", "total_floss", "total_mouthwash", "streak_days"]

X = df[features_simplificadas]
y = df["score"]

# Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Separar treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Criar a rede neural simplificada
model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(1)
])

# Compilar o modelo
model.compile(optimizer="adam", loss="mse", metrics=["mae"])

# Treinar o modelo
history = model.fit(X_train, y_train, epochs=200, batch_size=16, validation_data=(X_test, y_test), verbose=1)
```

### Avaliando o modelo

```python
# Avaliar o modelo
loss, mae = model.evaluate(X_test, y_test)
print(f"Erro médio absoluto (MAE): {mae:.2f}")
```

### Testando o modelo

```python
# Fazer previsões
y_pred = model.predict(X_test)

# Converter para inteiros
y_pred = y_pred.flatten().astype(int)

# Exibir as 5 primeiras previsões sem casas decimais
print("Exemplo de previsões:", y_pred[:5])
```
